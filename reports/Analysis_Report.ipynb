#### ii. Justification for Training-Test Split Ratio

An **80/20 training-test split** was used. This ratio is widely regarded as a practical standard in machine learning, providing sufficient data for model training (80%) while retaining enough held-out data (20%) for reliable generalisation estimation. Guyon (1997) recommends this ratio as it balances the bias-variance trade-off — too small a training set leads to underfitting, while too small a test set yields unreliable performance estimates. With 58,645 records, the 20% test set (~11,729 instances) is large enough to produce statistically meaningful evaluation results (James et al., 2013).

**References:**
- Guyon, I. (1997) 'A Scaling Law for the Validation-Set Training-Set Size Ratio', *AT&T Bell Laboratories Technical Report*.
- James, G., Witten, D., Hastie, T. and Tibshirani, R. (2013) *An Introduction to Statistical Learning: With Applications in R*. New York: Springer.

#### iii. Ensuring Fair Comparison — Same Test Instances and Stratification

> **Paste screenshot:** Code block line and output from Notebook 2 showing `train_test_split` with `random_state` and `stratify` parameters

**Parameters responsible for meeting both conditions:**
1. **`random_state=42`** — ensures reproducibility so that all models are tested on the **same test instances**
2. **`stratify=y`** — ensures the **class ratio** of "Approved" to "Rejected" is preserved in both training and test subsets

**Justification:** Setting `random_state` guarantees deterministic splits, meaning every model sees identical training and test partitions — this is essential for fair model comparison (Pedregosa et al., 2011). The `stratify` parameter ensures that the imbalanced class distribution (85.8% Rejected vs 14.2% Approved) is maintained in both subsets, preventing scenarios where one class is underrepresented in either partition. Without stratification, random splits on imbalanced data can produce test sets with significantly different class proportions, leading to misleading evaluation results (James et al., 2013). Together, these parameters ensure a rigorous and reproducible experimental design.

## Task (2) – Modelling: Build Predictive Regression Models

### a) Benefits of Using a Decision Tree Regressor

Decision Tree (DT) regressors offer several key benefits for predicting Maximum Loan Amount in this financial context. First, DTs are **highly interpretable** — they produce rule-based splits that finance analysts can understand and explain to clients, which is critical for regulatory compliance and transparency in loan decisions (Breiman et al., 1984). Second, DTs handle **non-linear relationships** between features and the target without requiring feature transformations, capturing complex interactions (e.g., between income and employment length) that linear models miss. Third, DTs provide **built-in feature importance rankings**, helping analysts identify which client attributes most influence the maximum loan offer. Fourth, DTs are **robust to outliers** and do not require feature scaling, reducing preprocessing requirements (James et al., 2013).

**References:**
- Breiman, L., Friedman, J.H., Olshen, R.A. and Stone, C.J. (1984) *Classification and Regression Trees*. Belmont, CA: Wadsworth.
- James, G., Witten, D., Hastie, T. and Tibshirani, R. (2013) *An Introduction to Statistical Learning: With Applications in R*. New York: Springer.

#### ii. Pruning Method Explanation

**1) Type of pruning used for DT-2:** **Pre-pruning** (also called early stopping) was applied by setting `max_depth=4`, which constrains the tree to grow no deeper than 4 levels during training. This limits complexity before the tree is fully grown.

**2) Benefits and Disadvantages:**

**Benefits:**
- **Interpretability:** A 4-level tree with only 16 leaves can be easily visualised and explained to bank analysts, supporting transparent loan decisions
- **Generalisation:** Pre-pruning reduces overfitting by preventing the tree from memorising training data, producing more reliable predictions on new clients
- **Computational Efficiency:** Smaller trees are faster to train and predict

**Disadvantages:**
- **Underfitting Risk:** Restricting depth to 4 may oversimplify the model, missing important patterns in the loan data such as nuanced interactions between income, employment length, and credit history
- **Information Loss:** Important splits that occur deeper in the tree are permanently excluded, potentially reducing accuracy for complex client profiles

### c) Concerns About Selected Performance Metrics

While MSE, MAE, and R² all favour DT-1 as the best model, there are notable concerns. **MSE is sensitive to outliers** — a small number of extreme loan values can disproportionately inflate MSE, potentially making a model appear worse than it truly is for typical predictions (Hastie et al., 2009). **R² can be misleadingly high** for overfitting models; DT-1's depth of 24 levels with 22,912 leaves may partially memorise training data rather than learning generalisable patterns, resulting in an inflated R² on the test set that may not hold for truly unseen data (James et al., 2013). **MAE treats all errors equally**, giving no extra penalty to dangerously large errors in extreme loan amounts — yet the success criteria specifically requires effective handling of extreme values. Additionally, none of these metrics capture **directional bias** — whether the model systematically over-predicts or under-predicts loan amounts, which has different financial implications for the bank. Over-prediction increases default risk, while under-prediction loses potential revenue.
