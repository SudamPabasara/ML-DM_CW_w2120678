## Snippet 2 (Cell 4): Imports and Global Settings
### Why this is usedLoads all modeling and plotting tools.
### How it worksImports regression and classification components, then applies reproducibility/display settings.
### How to change itRemove unused imports, adjust random seed and figure size.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, auc
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
import warnings
warnings.filterwarnings('ignore')
np.random.seed(42)
pd.set_option('display.max_columns', None)
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 8)
```

## Snippet 3 (Cell 5): Load Raw Dataset
### Why this is usedStarts both case studies from original CSV.
### How it worksReads loan_approval_data.csv and prints shape/columns.
### How to change itUpdate file path, delimiter, or encoding if source changes.

```python
df_full = pd.read_csv('loan_approval_data.csv')
print(df_full.shape)
print(list(df_full.columns))
```

## Snippet 4 (Cell 7): Classification Data Preparation
### Why this is usedRebuilds the same preprocessing pipeline used in Notebook 2.
### How it worksFills missing values, removes duplicates, prepares X/y, stratified split, imputation, scaling.
### How to change itModify dropped columns, imputation strategy, or test split ratio.

```python
df_clf = df_full.copy()
for c in df_clf.select_dtypes(include=['object']).columns:
    df_clf[c] = df_clf[c].fillna(df_clf[c].mode()[0])
for c in df_clf.select_dtypes(include=[np.number]).columns:
    df_clf[c] = df_clf[c].fillna(df_clf[c].mean())
df_clf.drop_duplicates(inplace=True)
X_approval = df_clf.drop(['id', 'loan_approval_status', 'max_allowed_loan', 'home_ownership', 'loan_intent', 'payment_default_on_file'], axis=1)
y_approval = df_clf['loan_approval_status']
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_approval, y_approval, test_size=0.2, random_state=42, stratify=y_approval)
imp_clf = SimpleImputer(strategy='mean')
X_train_clf = pd.DataFrame(imp_clf.fit_transform(X_train_clf), columns=X_approval.columns)
X_test_clf = pd.DataFrame(imp_clf.transform(X_test_clf), columns=X_approval.columns)
scaler_clf = StandardScaler()
X_train_clf_scaled = scaler_clf.fit_transform(X_train_clf)
X_test_clf_scaled = scaler_clf.transform(X_test_clf)
```

## Snippet 5 (Cell 9): Train Base Learners and Ensemble
### Why this is usedBuilds two-level ensemble workflow for classification.
### How it worksTrains NB/LR/KNN, selects top two by F1, builds soft-voting ensemble, evaluates metrics.
### How to change itSelect base models manually or change selection metric from F1 to ROC-AUC.

```python
nb = GaussianNB().fit(X_train_clf_scaled, y_train_clf)
lr = LogisticRegression(max_iter=1000, random_state=42).fit(X_train_clf_scaled, y_train_clf)
knn = KNeighborsClassifier(n_neighbors=5).fit(X_train_clf_scaled, y_train_clf)
base_models = {'Naive Bayes': nb, 'Logistic Regression': lr, 'KNN': knn}
rows = []
for n, m in base_models.items():
    rows.append({'Model': n, 'F1': f1_score(y_test_clf, m.predict(X_test_clf_scaled), zero_division=0)})
base_df = pd.DataFrame(rows).sort_values('F1', ascending=False)
best_two = base_df.head(2)['Model'].tolist()
voting_clf = VotingClassifier(estimators=[('m1', base_models[best_two[0]]), ('m2', base_models[best_two[1]])], voting='soft').fit(X_train_clf_scaled, y_train_clf)
y_pred_ens = voting_clf.predict(X_test_clf_scaled)
y_proba_ens = voting_clf.predict_proba(X_test_clf_scaled)[:, 1]
print(best_two)
print('F1:', f1_score(y_test_clf, y_pred_ens), 'ROC-AUC:', roc_auc_score(y_test_clf, y_proba_ens))
```

## Snippet 6 (Cell 10): Ensemble and Base Visualizations
### Why this is usedCompares confusion matrices and ROC curves visually.
### How it worksLoops through selected base models plus ensemble and plots diagnostics.
### How to change itSave figures to custom folder or add precision-recall curve.

```python
models_to_plot = {best_two[0]: base_models[best_two[0]], best_two[1]: base_models[best_two[1]], 'Voting Ensemble': voting_clf}
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
for i, (n, m) in enumerate(models_to_plot.items()):
    cm = confusion_matrix(y_test_clf, m.predict(X_test_clf_scaled))
    sns.heatmap(cm, annot=True, fmt='d', ax=axes[0, i], cmap='Blues')
    axes[0, i].set_title(n)
    p = m.predict_proba(X_test_clf_scaled)[:, 1]
    fpr, tpr, _ = roc_curve(y_test_clf, p)
    axes[1, i].plot(fpr, tpr, label=f'AUC={auc(fpr, tpr):.3f}')
    axes[1, i].plot([0, 1], [0, 1], 'r--')
    axes[1, i].legend()
plt.tight_layout()
```

## Snippet 7 (Cell 11): Ensemble vs Base Comparison Table/Chart
### Why this is usedSummarizes performance differences across models.
### How it worksCalculates five metrics per model, stores in dataframe, plots grouped bars.
### How to change itAdd calibration metric (Brier score) or reorder metrics.

```python
comp = []
for n, m in models_to_plot.items():
    y = m.predict(X_test_clf_scaled)
    p = m.predict_proba(X_test_clf_scaled)[:, 1]
    comp.append({'Model': n, 'Accuracy': accuracy_score(y_test_clf, y), 'Precision': precision_score(y_test_clf, y, zero_division=0), 'Recall': recall_score(y_test_clf, y, zero_division=0), 'F1': f1_score(y_test_clf, y, zero_division=0), 'ROC-AUC': roc_auc_score(y_test_clf, p)})
comp_df = pd.DataFrame(comp)
print(comp_df.to_string(index=False))
```

## Snippet 8 (Cell 13): Save Ensemble Artifacts
### Why this is usedReuses trained ensemble and selected base models later.
### How it worksCreates models folder and dumps joblib files.
### How to change itSave in workspace folder instead of parent folder to avoid path confusion.

```python
import os
os.makedirs('models', exist_ok=True)
joblib.dump(voting_clf, 'models/ensemble_clf.joblib')
joblib.dump(base_models[best_two[0]], 'models/' + best_two[0].replace(' ', '_') + '_base.joblib')
joblib.dump(base_models[best_two[1]], 'models/' + best_two[1].replace(' ', '_') + '_base.joblib')
```

## Snippet 9 (Cell 17): Regression Dataset Preparation and Target Analysis
### Why this is usedDefines supervised regression problem for max_allowed_loan.
### How it worksFilters known targets, selects features, prints stats, plots distribution.
### How to change itInclude engineered ratio features or winsorize target outliers.

```python
df_raw = pd.read_csv('loan_approval_data.csv')
df_reg = df_raw[df_raw['max_allowed_loan'] > 0].copy()
drop_cols = ['id', 'loan_approval_status', 'max_allowed_loan', 'home_ownership', 'loan_intent', 'payment_default_on_file']
X_lending = df_reg.drop(drop_cols, axis=1)
y_lending = df_reg['max_allowed_loan']
print(X_lending.shape, y_lending.shape)
print(y_lending.describe())
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].hist(y_lending, bins=40)
ax[1].boxplot(y_lending)
plt.tight_layout()
```

## Snippet 10 (Cell 19): Benefits List for Decision Tree Regression
### Why this is usedDocuments algorithm suitability for coursework narrative.
### How it worksStores benefits in list and prints each line.
### How to change itAdd dataset-specific evidence under each benefit.

```python
benefits = [
    'Interpretability',
    'Non-linear pattern capture',
    'Feature importance visibility',
    'No strict scaling requirement'
]
for b in benefits:
    print('-', b)
```

## Snippet 11 (Cell 21): Train/Test Split and Imputation for Regression
### Why this is usedCreates robust train/test sets and handles missing values.
### How it worksSplits data then uses mean imputation fit on training set.
### How to change itUse median imputation for skewed numeric features.

```python
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_lending, y_lending, test_size=0.2, random_state=42)
imp_reg = SimpleImputer(strategy='mean')
X_train_reg = pd.DataFrame(imp_reg.fit_transform(X_train_reg), columns=X_lending.columns)
X_test_reg = pd.DataFrame(imp_reg.transform(X_test_reg), columns=X_lending.columns)
print(X_train_reg.shape, X_test_reg.shape)
```

## Snippet 12 (Cell 22): Build DT-1 Fully Grown
### Why this is usedBaseline high-capacity tree.
### How it worksFits DecisionTreeRegressor without max_depth.
### How to change itAdd min_samples_leaf to reduce overfitting.

```python
dt1 = DecisionTreeRegressor(random_state=42)
dt1.fit(X_train_reg, y_train_reg)
print(dt1.get_depth(), dt1.get_n_leaves())
```

## Snippet 13 (Cell 23): Build DT-2 Pruned
### Why this is usedLower-complexity tree for better generalization.
### How it worksUses max_depth=4 as pre-pruning.
### How to change itTune max_depth with grid search.

```python
dt2 = DecisionTreeRegressor(max_depth=4, random_state=42)
dt2.fit(X_train_reg, y_train_reg)
print(dt2.get_depth(), dt2.get_n_leaves())
```

## Snippet 14 (Cell 25): DT-1 Visualization and Feature Importance
### Why this is usedInterprets decision logic and influential predictors.
### How it worksPrints structure stats, computes feature importances, plots top levels of tree.
### How to change itIncrease/decrease max_depth shown in plot_tree for readability.

```python
fi1 = pd.DataFrame({'Feature': X_train_reg.columns, 'Importance': dt1.feature_importances_}).sort_values('Importance', ascending=False)
print(fi1.to_string(index=False))
fig, ax = plt.subplots(figsize=(18, 8))
plot_tree(dt1, feature_names=list(X_train_reg.columns), filled=True, rounded=True, max_depth=3, ax=ax)
plt.tight_layout()
```

## Snippet 15 (Cell 26): DT-2 Visualization and Importance Comparison
### Why this is usedCompares pruned and full trees side by side.
### How it worksVisualizes DT-2 and plots horizontal bar charts for both feature-importance profiles.
### How to change itNormalize feature importances for easier visual comparison across models.

```python
fi2 = pd.DataFrame({'Feature': X_train_reg.columns, 'Importance': dt2.feature_importances_}).sort_values('Importance', ascending=False)
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
ax[0].barh(fi1.sort_values('Importance')['Feature'], fi1.sort_values('Importance')['Importance'])
ax[0].set_title('DT-1 Importances')
ax[1].barh(fi2.sort_values('Importance')['Feature'], fi2.sort_values('Importance')['Importance'])
ax[1].set_title('DT-2 Importances')
plt.tight_layout()
```

## Snippet 16 (Cell 28): Evaluation Metric Selection Table
### Why this is usedJustifies MSE/MAE/R2 as key regression metrics.
### How it worksCreates a dataframe with USE/DO NOT USE reasoning.
### How to change itAdd RMSE for direct comparison in target units.

```python
metrics_table = pd.DataFrame([
    {'Metric': 'MSE', 'Use': 'USE', 'Reason': 'Penalizes large errors'},
    {'Metric': 'MAE', 'Use': 'USE', 'Reason': 'Error in dollars'},
    {'Metric': 'R2', 'Use': 'USE', 'Reason': 'Variance explained'}
])
print(metrics_table.to_string(index=False))
```

## Snippet 17 (Cell 31): Compute Metrics and Diagnostic Plots
### Why this is usedQuantifies predictive quality and error behavior.
### How it worksPredicts DT-1/DT-2, computes MSE/MAE/R2, plots actual-vs-predicted and residual histograms.
### How to change itAdd residual vs fitted plots for heteroscedasticity checks.

```python
pred1 = dt1.predict(X_test_reg)
pred2 = dt2.predict(X_test_reg)
mse1, mae1, r21 = mean_squared_error(y_test_reg, pred1), mean_absolute_error(y_test_reg, pred1), r2_score(y_test_reg, pred1)
mse2, mae2, r22 = mean_squared_error(y_test_reg, pred2), mean_absolute_error(y_test_reg, pred2), r2_score(y_test_reg, pred2)
res = pd.DataFrame([{'Model':'DT-1','MSE':mse1,'MAE':mae1,'R2':r21},{'Model':'DT-2','MSE':mse2,'MAE':mae2,'R2':r22}])
print(res.to_string(index=False))
```

## Snippet 18 (Cell 32): Choose Best Regressor
### Why this is usedSelects deployment candidate using chosen metric logic.
### How it worksCompares MSE and assigns best model object and summary values.
### How to change itUse weighted score combining MSE and MAE.

```python
if mse1 < mse2:
    best_model, best_reg = 'DT-1', dt1
else:
    best_model, best_reg = 'DT-2', dt2
print('Best model:', best_model)
```
